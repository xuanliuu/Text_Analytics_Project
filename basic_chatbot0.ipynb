```python
from chatterbot.trainers import ChatterBotCorpusTrainer
chatbot = ChatBot('learner', preprocessors=[
        'chatterbot.preprocessors.clean_whitespace'], logic_adapters=[
        {
            "import_path": "chatterbot.logic.BestMatch",
            "statement_comparison_function": chatterbot.comparison.LevenshteinDistance,
            "response_selection_method": chatterbot.response_selection.get_first_response
        }
    ], filters=[filters.get_recent_repeated_responses])
```

```python
trainer = ListTrainer(chatbot)
```

```python
trainer.train("chatterbot.corpus.english.greetings")
```

```python
trainer.train(['How can i get intership?',
              'We provide 3-weeks intership',
              'What kind of experiences can I get through the Innovation Center?',
              'An internship',
               'Can I get career guidance through the Innovation Center?',
              'Yes, the Innovation Center provides career guidance',
              'Who does the Innovation Center partner with?',
              'Both domestic and international companies'])
```
