```python
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
import chatterbot
import dash
import dash_table
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output, State
import numpy as np
import pandas as pd
import warnings
import dash_bootstrap_components as dbc
```

```python
chatbot = ChatBot('learner')
```

```python
trainer = ListTrainer(chatbot)
```

```python
trainer.train(['How can i get intership?',
              'We provide 3-weeks intership',
              'What kind of experiences can I get through the Innovation Center?',
              'An internship',
               'Can I get career guidance through the Innovation Center?',
              'Yes, the Innovation Center provides career guidance',
              'Who does the Innovation Center partner with?',
              'Both domestic and international companies'])
```
